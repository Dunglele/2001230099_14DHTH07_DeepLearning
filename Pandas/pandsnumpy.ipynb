### Câu 1: Kiểm tra mảng khác 0 (np.all)
Tạo mảng 3x3 với các số nguyên ngẫu nhiên trong khoảng [0, 20]. Kiểm tra xem tất cả các phần tử có khác 0 hay không.

```python
arr1 = np.random.randint(0, 21, size=(3, 3))
print("Mảng 1:\n", arr1)
print("Tất cả phần tử khác 0:", np.all(arr1 != 0))
```

### Câu 2: Kiểm tra tồn tại phần tử khác 0 (np.any)
Tạo mảng 3x3 với các số nguyên ngẫu nhiên trong khoảng [0, 20]. Kiểm tra xem có tồn tại ít nhất một phần tử khác 0 hay không.

```python
arr2 = np.random.randint(0, 21, size=(3, 3))
print("Mảng 2:\n", arr2)
print("Có ít nhất một phần tử khác 0:", np.any(arr2 != 0))
```

### Câu 3: So sánh mảng
Sử dụng `np.greater`, `np.greater_equal`, `np.less`, `np.less_equal` để so sánh 2 mảng.

```python
x = np.array([1, 2, 3])
y = np.array([2, 2, 2])
print("x > y:", np.greater(x, y))
print("x >= y:", np.greater_equal(x, y))
print("x < y:", np.less(x, y))
print("x <= y:", np.less_equal(x, y))
```

### Câu 8: Đổi dấu trong khoảng
Tạo mảng 20 phần tử trong khoảng [0, 20] và đổi dấu những số nằm trong khoảng [9, 15].

```python
arr8 = np.arange(21)
arr8[(arr8 >= 9) & (arr8 <= 15)] *= -1
print(arr8)
```

### Câu 12: Mảng 3x3x3
Tạo mảng 3x3x3 và tính tổng theo mỗi dòng, mỗi cột.

```python
arr12 = np.arange(27).reshape((3, 3, 3))
print("Mảng 3D:\n", arr12)
print("Tổng theo cột (axis 1):\n", np.sum(arr12, axis=1))
print("Tổng theo dòng (axis 2):\n", np.sum(arr12, axis=2))
```

## BÀI TẬP PANDAS

### Câu 15: Các phép toán trên Series
Cộng, trừ, nhân, chia 2 Series: [2, 4, 6, 8, 10] và [1, 3, 5, 7, 10].

```python
s1 = pd.Series([2, 4, 6, 8, 10])
s2 = pd.Series([1, 3, 5, 7, 10])
print("Cộng:\n", s1 + s2)
print("Trừ:\n", s1 - s2)
print("Nhân:\n", s1 * s2)
print("Chia:\n", s1 / s2)
```

### Câu 16: Cột DataFrame thành Series

```python
df16 = pd.DataFrame({'col1': [1, 2, 3], 'col2': [4, 5, 6]})
series_col = df16.iloc[:, 0]
print(series_col)
print(type(series_col))
```

### Câu 17: Sắp xếp DataFrame

### Câu 18: Thay đổi index
Thay đổi index của một Series.

```python
s18 = pd.Series([1, 2, 3, 4, 5], index=['A', 'B', 'C', 'D', 'E'])
print("Gốc:\n", s18)
s18 = s18.reindex(['B', 'A', 'C', 'D', 'E'])
print("Sau khi reindex:\n", s18)
```

### Câu 19: Tìm phần tử không thuộc Series khác
Tìm những phần tử trong Series x mà không có trong Series y.

```python
sr1 = pd.Series([1, 2, 3, 4, 5])
sr2 = pd.Series([2, 4, 6, 8, 10])
result = sr1[~sr1.isin(sr2)]
print(result)
```

### Câu 20: Hội và Giao của mảng
Minh họa `np.union1d` và `np.intersect1d` trên 2 Series ngẫu nhiên 20 phần tử.

```python
sx = pd.Series(np.random.randint(0, 10, 20))
sy = pd.Series(np.random.randint(0, 10, 20))
print("Hội:", np.union1d(sx, sy))
print("Giao:", np.intersect1d(sx, sy))
```

### Câu 21: Thêm Series (Dọc và Ngang)
Thêm series y vào series x theo chiều dọc và chiều ngang.

```python
ser_x = pd.Series([0, 1, 2])
ser_y = pd.Series(['p', 'q', 'r'])
print("Chiều dọc (Vertical):\n", pd.concat([ser_x, ser_y], axis=0))
print("Chiều ngang (Horizontal):\n", pd.concat([ser_x, ser_y], axis=1))
```

### Câu 22: Tạo DataFrame exam_data

```python
exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(exam_data, index=labels)
print(df)
```
